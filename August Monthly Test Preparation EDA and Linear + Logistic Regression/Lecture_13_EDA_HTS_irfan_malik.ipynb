# **Handling Outliers With IQR Dropping the entire rows having outliers**

```python
Q1 = df["Engine HP"].quantile(0.25)

Q3 = df["Engine HP"].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR

upper_bound = Q3 + 1.5 * IQR

df["Engine HP"].dtype

df = df[(df["Engine HP"] >= lower_bound) & (df["Engine HP"] <= upper_bound)]
```
